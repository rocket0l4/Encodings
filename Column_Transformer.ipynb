```python
import numpy as np
import pandas as pd
```

```python
df = df.drop(['Unit_Price','Revenue','Profit','Unit_Cost','Order_Quantity','Age_Group','State','Sub_Category'],axis=1)
```

```python
df = df.drop(['Date','Day','Month'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Cost']),df['Cost'],
                                                 test_size=0.2)
```

#Simply Appliying encoding one by one on columns.

```python
# Odinal Encoding Product_Category
pr = OrdinalEncoder(categories=[['Bikes','Clothing','Accessories']])
X_train_prod =pr.fit_transform(X_train[['Product_Category']])

# For test data
X_test_prod = pr.fit_transform(X_test[['Product_Category']])

X_train_prod.shape
```

```python
# OneHotEncoding -> Customer_Gender,Country
ohe = OneHotEncoder(sparse= False,)
X_train_gen_cnty = ohe.fit_transform(X_train[['Customer_Gender','Country']])

# For test data
X_test_gen_cnty = ohe.fit_transform(X_test[['Customer_Gender','Country']])

X_train_gen_cnty.shape
```

```python
# Extracting Age,year,product.
X_train_age_yr_prod = X_train.drop(columns=['Customer_Gender','Country','Product_Category']).values

# From Test Data
X_test_age_yr_prod = X_test.drop(columns= ['Customer_Gender','Country','Product_Category']).values


X_train_age_yr_prod.shape
```

#Concatenate

```python
# for train
X_train_transformed = np.concatenate((X_train_age_yr_prod,X_train_gen_cnty,X_train_prod),axis=1)

# For test
X_test_transformed = np.concatenate((X_test_age_yr_prod,X_test_gen_cnty,X_test_prod),axis=1)

X_train_transformed.shape
```

# Column Transformer
- using of the column transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('mob1',OrdinalEncoder(categories=[['Bikes','Clothing','Accessories']]),['Product_Category']),
    ('mob2',OneHotEncoder(sparse= False),['Customer_Gender','Country'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train)
```
